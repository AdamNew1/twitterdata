```python
weather = {'rain' : ['rain', 'pouring', 'drizzle', 'drizzly', 'deluge', 'storm'], 'fogmist' : ['fog', 'mist'], 'snowhail' : ['snow', 'hail', 'blizzard', 'frost', 'SNOW'], 'sunny' : ['sunny', 'sunlight', 'sunshine'], 'cloudy' : ['gloomy', 'overcast', 'cloud', 'dark sky', 'grey sky'], 'thunder' : ['thunder', 'lightning']}
```

```python
def weatherf(cell_value):
    return {'rain' : sum({v for k, v in ((s, cell_value.count(s)) for s in rain) if v}), 'sunny' : sum({v for k, v in ((s, cell_value.count(s)) for s in sunny) if v}), 'cloudy' : sum({v for k, v in ((s, cell_value.count(s)) for s in cloudy) if v}), 'fog/mist' : sum({v for k, v in ((s, cell_value.count(s)) for s in fogmist) if v}), 'snow/hail' : sum({v for k, v in ((s, cell_value.count(s)) for s in snowhail) if v}), 'thunder' : sum({v for k, v in ((s, cell_value.count(s)) for s in thunder) if v})}
     
```

```python
uberfiltered['weathercount'] = uberfiltered['text'].apply(weatherf)
```

```python
uberfiltered
```

```python
def definitions(cell):
    highest = max(cell.values())
    if highest > 0:
        dictlist = [k for k,v in cell.items() if v == highest]
        return ' , '.join(str(d) for d in dictlist)
```

```python
uberfiltered['tweetweather'] = uberfiltered['weathercount'].apply(definitions)
```
