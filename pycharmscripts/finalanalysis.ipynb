```python
fullpos = pd.concat([timpos,zarpos]).reset_index(drop=True)
```

```python
fullpos['current'] = pd.to_datetime(fullpos['timestamp_ms'], utc = True, unit = 'ms')
```

```python
fullpos['shortened'] = fullpos['current'].map(lambda t: t.strftime('%Y-%m-%d %H:00:00'))
```

```python
time1 = metoffice[['Forecast Time', 'Degrees Celsius', 'Weather']].drop_duplicates()
time1
```

```python
fuckedup = pd.merge(time1, tweettime, on = 'key', how = 'outer')
```

```python
fuckedup['difference'] = pd.to_datetime(fuckedup['Forecast Time']) - pd.to_datetime(fuckedup['shortened'])
```
